The size of all images must be the same and should be converted to 4D arrays so that they can be used as an input for the convolutional neural network. Also, we need to create the labels for both types of images to be able to train and test the model.

Let's do the same for the training data first and then we will use the same code for the test data as well.

```python
#store path of the "train" folder
train_dir = '/content/cell_images/train'

#ensure images have the same size
IMG_SIZE = 64

#create empty lists to store train images and labels
train_images = []
train_labels = []
```

```python
#function for "parasitized" and "uninfected" folders in the train set
for folder_name in ['/parasitized/','/uninfected/']:


    #foder path
    path = os.listdir(train_dir + folder_name)

    #For loop to assign labels for images in the folders
    for i, image_name in enumerate(path):

      try:
        #Open each image using path of that image
        image = Image.open(train_dir + folder_name + image_name)


        #Resize each image to 64 x 64
        image = image.resize((IMG_SIZE, IMG_SIZE))

        #Convert images to arrays and append to empty list
        train_images.append(np.array(image))

        # Creating labels for parasitized and uninfected images
        if folder_name == '/parasitized/':
          train_labels.append(1)
        elif folder_name == '/uninfected/':
          train_labels.append(0)

      except Exception:

        pass
```

```python
#Convert image and label lists to arrays
train_images = np.array(train_images)

train_labels = np.array(train_labels)
```

```python
#store path of the "test" folder
test_dir = '/content/cell_images/test'

#ensure images in the folder are the same size
IMG_SIZE = 64

#create empty list to store test images and label
test_images = []
test_labels = []
```

```python
#function for "parasitized" and "uninfected" folders in the test set

for folder_name in ['/parasitized/','/uninfected/']:


    #Folder path isted
    path = os.listdir(test_dir + folder_name)

    #For loop to assign labels for images in the folders
    for i, image_name in enumerate(path):

      try:
        #Open each image using path of that image
        image = Image.open(test_dir + folder_name + image_name)


        #Resize each image to 64 x 64
        image = image.resize((IMG_SIZE, IMG_SIZE))

        #Converting images to arrays and append to empty lists
        test_images.append(np.array(image))

        #Create labesl for parasitized and uninfected images
        if folder_name == '/parasitized/':
          test_labels.append(1)
        elif folder_name == '/uninfected/':
          test_labels.append(0)

      except Exception:

        pass
```

```python
#Convert lists of test images and labels to arrays
test_images = np.array(test_images)

test_labels = np.array(test_labels)
```

###<b> Importing the required libraries for building and training our Model

```python
#import required libraries from tensorfow and keras to build model
import tensorflow as tf
import keras
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, Conv2D, MaxPooling2D, BatchNormalization, Activation, Input, LeakyReLU, Dropout, Flatten
from tensorflow.keras import backend
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import Adam, SGD, RMSprop
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras import losses, optimizers
from tensorflow.keras.preprocessing.image import load_img
from google.colab.patches import cv2_imshow
```

```python
#clear backend
backend.clear_session()
```

```python
#import random library and set random seed
import random

seed = 42
random.seed()
np.random.seed(seed)
tf.random.set_seed(seed)
```

###<b> Building the model

```python
#Initialize sequential model
model = Sequential()

#First convolutional layer w/ 32 filters, kernel size of 2, padding = same, relu activation
model.add(Conv2D(filters = 32, kernel_size = 2, padding = "same", activation = "relu", input_shape = (64, 64, 3)))

#Max pooling layer to reduce the size of output
model.add(MaxPooling2D(pool_size = 2))

#Dropout to randomly switch of 20% of neurons to reduce overfitting
model.add(Dropout(0.2))

#Second convolutional layer
model.add(Conv2D(filters = 32, kernel_size = 2, padding = "same", activation = "relu"))

#Maxpooling on 2nd convolutional layer
model.add(MaxPooling2D(pool_size = 2))

#Additional dropout layer
model.add(Dropout(0.2))

#Third convolutional layer
model.add(Conv2D(filters = 32, kernel_size = 2, padding = "same", activation = "relu"))

#Maxpooling on 3rd layer
model.add(MaxPooling2D(pool_size = 2))

#Additional dropout layer
model.add(Dropout(0.2))

#Flatten output of convolutional layer to prep it for dense connections
model.add(Flatten())

#Add fully connected dense layer w/ 512 neurons
model.add(Dense(512, activation = "relu"))

#Additional dropout layer on 40% of neurons
model.add(Dropout(0.4))

#Add output layer w/ 2 neurons for 2 classes and 'softmax' activation
model.add(Dense(2, activation = "softmax"))

#print model summary
model.summary()
```

###<b> Compiling the model

###<b> Model 2 with Batch Normalization

###<b> Building the Model

```python
#Initialize sequential model 2
model2 = Sequential()

#First convolutional layer with 32 filters, kernel size of 3x3, padding = same, input shape of 64 x 64
model2.add(Conv2D(32, (3, 3), input_shape = (64, 64, 3), padding = 'same'))

#LeakyRelu activation function with alpha of 0.1
model2.add(LeakyReLU(alpha = 0.1))

#Second convolutional layer
model2.add(Conv2D(32, (3, 3), padding = 'same'))

#LeakyRelu activation function with alpha of 0.1
model2.add(LeakyReLU(alpha = 0.1))

#Max pooling to reduce size of the output from 2nd convolutional layer
model2.add(MaxPooling2D(pool_size = (2,2)))

#Dropout of 20% added to 2nd convolutional layer
model2.add(Dropout(0.2))

#Third convolutional layer
model2.add(Conv2D(32, (3, 3), padding = 'same'))

#LeakyRelu activation function w/ alpha of 0.1
model2.add(LeakyReLU(alpha = 0.1))

#Fourth convolutional layer
model2.add(Conv2D(32, (3, 3), padding = 'same'))

#LeakyRelu activation function w/ alpha of 0.1
model2.add(LeakyReLU(alpha = 0.1))

#Maxpooing layer to reduce size of output from 4th layer
model2.add(MaxPooling2D(pool_size = (2,2)))

#Batch normalization layer to normalize inputs and prevent vanishing or exploding gradients
model2.add(BatchNormalization())

#dropout of 20% added to 4th convolutional layer
model2.add(Dropout(0.2))

#Flatten output of convolutional layer to prep it for dense connections
model2.add(Flatten())

#Add fully connected dense layer w/ 512 neurons
model2.add(Dense(512))

#LeakyRelu activation function w/ alpha of 0.1
model2.add(LeakyReLU(alpha = 0.1))

#Dropout added to 40% of neurons
model2.add(Dropout(0.4))

#Fully connected dense layer w/ 2 neurons
model2.add(Dense(2, activation = "softmax"))

#Print model 2 summary
model2.summary()
```

```python
#import VGG16 libraries
from tensorflow.keras.applications.vgg16 import VGG16
from tensorflow.keras import Model
```

```python
#get VGG pretrained model, incude_top = False to omit the last 3 dense layers, input original shape of 64 x 64
vgg = VGG16(weights = 'imagenet',
            include_top= False,
            input_shape= (64, 64, 3))

#check summary of the model
vgg.summary()
```

```python
#use block 3 layer to get output as the transfer layer in VGG model
transfer_layer = vgg.get_layer('block3_pool')

#keep the same parameters of VGG
vgg.trainable = False
```

```python
# Flatten the output from the 3rd block of the VGG16 model
x = Flatten()(transfer_layer.output)

#intermediate layer
x = Dense(256, activation = 'relu')(x)

# Similarly add a dense layer with 128 neurons
x = Dense(128, activation = 'relu')(x)

#dropout incuded on 30% of neurons
x = Dropout(0.3)(x)

#Add another dense layer
x = Dense(63, activation = 'relu')(x)

#Add BatchNormalization layer
x = BatchNormalization()(x)

#output layer w/ 2 neurons for 2 classes
pred = Dense(2, activation = 'softmax')(x)

#Initializing Model4
model4 = Model(vgg.input, pred)
```

###<b>Compiling the model

```python
#model 4 compiled with smaller learning rate in optimizer because it is pretrained
model4.compile(loss = 'binary_crossentropy', optimizer = tf.keras.optimizers.Adam(learning_rate=0.0005), metrics = ['accuracy'])
```
